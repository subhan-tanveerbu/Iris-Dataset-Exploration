Task 1: Exploring and Visualizing a Simple Dataset

Importing Required Libraries

```python
# Import pandas for data manipulation
import pandas as pd

# Import seaborn for dataset and visualization
import seaborn as sns

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# This ensures plots are displayed inside the notebook
%matplotlib inline



```

Loading the dataset

```python
df = sns.load_dataset("iris") # Load the built-in Iris dataset from seaborn
```

Scatter Plot (Relationship Between Features)

```python
# Create figure
plt.figure(figsize=(8,6))

# Scatter plot
sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    data=df
)

# Book-style figure numbering
plt.title("Figure 1.1: Sepal Length vs Petal Length by Species")

# Axis labels
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")

plt.legend(title="Species")
plt.show()
```

Histograms (Distribution of Features)

```python
# Define colors
colors = ["red", "green", "blue", "purple"]

# Select numerical columns
numerical_cols = df.select_dtypes(include=['float64']).columns

plt.figure(figsize=(10,8))

for i, col in enumerate(numerical_cols):

    plt.subplot(2, 2, i+1)

    plt.hist(df[col], color=colors[i], edgecolor="black")

    # Book-style numbering (1.x)
    plt.title(f"Figure 1.{i+2}: {col.replace('_',' ').title()} Distribution")

    plt.xlabel(col.replace('_',' ').title() + " (cm)")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

Boxplot (Outlier Detection)

```python
plt.figure(figsize=(8,6))

sns.boxplot(data=df)

plt.title("Figure 1.6: Boxplot of Iris Features")

plt.xlabel("Features")
plt.ylabel("Measurement (cm)")

plt.show()
```

Pairplot (Advanced Visualization)

```python
pairplot = sns.pairplot(df, hue="species")

# Book-style title
pairplot.fig.suptitle("Figure 1.7: Pairwise Feature Relationships by Species", y=1.02)

plt.show()
```
